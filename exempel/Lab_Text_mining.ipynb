# **Linguistic Features:**

Text: The original word text.

Lemma: The base form of the word.

POS: The simple POS part-of-speech tag.

Tag: The detailed part-of-speech tag. ("DT" for determiner, "VBZ" for verb, "NN" for noun)

Dep: Syntactic dependency, i.e. the relation between tokens.

Shape: The word shape – capitalization, punctuation, digits.

is alpha: Is the token an alphabetic character?

is stop: Is the token part of a stop list, i.e. the most common words of the language?

```python
# Lets calculate the Polarity of the Reviews
def get_polarity(text):
    textblob = TextBlob(str(text))
    pol = textblob.sentiment.polarity
    print(pol)
    if(pol==0):
        return "Neutral"
    elif(pol>0.1 and pol<=1):
        return "Positive"
    elif(pol>-1 and pol<=-0.1):
        return "Negative"
    
data['polarity'] = data['cleaned_data'].apply(get_polarity)
```

# Classification

```python
from textblob import TextBlob
import matplotlib.pyplot as plt
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re

data = pd.read_csv('test_data_class.tsv', delimiter = '\t')

data
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.dropna(inplace=True)

data
```

```python
r=0
blanks = []  # start with an empty list

for i in data['Review']:
    if i.isspace():
        blanks.append(r)

    r=r+1
        
blanks
```

```python
data
```

```python
def clean_data(text):
    return ' '.join(
        re.sub(
            r"(@[a-zA-Z0-9]+)|([^0-9A-Za-z])|(https://[\w.]+/[\w]+)",
            " ",
            text
        ).split()
    )

data['cleaned_data'] = data['Review'].apply(clean_data)

def drop_numbers(list_text):
    list_text_new = []
    for i in list_text:
        if not re.search(r'\d', i):
            list_text_new.append(i)
    return ''.join(list_text_new)

data['cleaned_data'] = data['cleaned_data'].apply(drop_numbers)
```

```python
data
```

```python
# Classification begins here

from sklearn.model_selection import train_test_split

X = data['cleaned_data']
y = data['Label']

# change your train and test size to check accuracy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Pipeline

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

# Creating classifier With data of vectors TFIDF
# Term frequency
text_clf_lsvc = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),
])

```

```python
# Training 
text_clf_lsvc.fit(X_train, y_train)
```

```python
# Testing
predictions = text_clf_lsvc.predict(X_test)
```

```python
from sklearn import metrics

# Print the overall accuracy
print(metrics.accuracy_score(y_test,predictions))
```

# Embeddings and semantic space

```python
# Install if needed
# pip install sentence-transformers scikit-learn

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example sentences
sentences = [
    "I love artificial intelligence",
    "Machine learning is fascinating",
    "Pizza tastes delicious",
    "Deep learning is a subset of AI"
]

# Generate embeddings
# converted into a 384-dimensional vector
embeddings = model.encode(sentences)
print("Embedding vector size:", len(embeddings[0]))

# Convert to dataframe for display
df = pd.DataFrame(embeddings)

print(df.head())

# Similarity comparison with first sentence
similarity = cosine_similarity([embeddings[0]], embeddings)

print("\nSimilarity Scores:")
for sentence, score in zip(sentences, similarity[0]):
    print(f"{sentence}: {score:.3f}")
```

# Sentiment analysis using pre-trained model

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment"
    #model="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
)

reviews = [
    "The movie was amazing",
    "I hated this film",
    "The product is okay",
    "It is average",
    "Fantastic acting and story",
    "Terrible experience"
]

results = classifier(reviews)

for review, result in zip(reviews, results):
    print("Review:", review)
    print("Prediction:", result["label"])
    print("Score:", round(result["score"], 3))
    print("-" * 50)
```
